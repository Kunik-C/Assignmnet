# **1. What is the difference between a function and a method in Python?**

**Ans 1.**

In Python, both functions and methods are blocks of code that perform specific tasks, but they differ primarily in how they are called and the context in which they operate.

    1)  Context: Functions are called independently, while methods are called on objects or classes.
    2)  Association: Functions are not tied to a class or object, but methods are.
    3)  First Argument: Methods typically take self as their first argument, referring to the object instance, while functions don't require this unless manually passed.


```python
#eample of function
def my_function():
    print("This is a function")

my_function()
```

```python
#example of method
class MyClass:
    def my_method(self):
        print("This is a method")

obj = MyClass()
obj.my_method()
```

# **5. What are iterators in Python and how do they differ from iterables?**

Ans 5.

In Python, iterators and iterables are closely related concepts, but they serve different purposes.

1. Iterable

    An iterable is any Python object that can be looped over (iterated). Examples include lists, tuples, strings, dictionaries, and sets.

    An iterable has the special method __iter__() that returns an iterator.

2. Iterator

    An iterator is an object that produces values one at a time when iterated over. It remembers the current position in the sequence and returns the next value upon each call to its __next__() method.

    An iterator is obtained by calling iter() on an iterable, and it implements two methods:

    __iter__(): Returns the iterator object itself.
    
    __next__(): Returns the next value from the iterator.
    
     When there are no more items to return, it raises a StopIteration exception.

```python
#example of iterable
# List is iterable
my_list = [1, 2, 3]
for item in my_list:
    print(item)

```

```python
#example of iterator
my_list = [1, 2, 3]
iterator = iter(my_list)

print(next(iterator))
print(next(iterator))
print(next(iterator))
```

# **6. Explain the concept of generators in Python and how they are defined.**

Ans 6.

In Python, generators are a type of iterable that allow you to iterate over a sequence of values lazily (i.e., one at a time, on-demand) without storing the entire sequence in memory. Generators are useful when dealing with large datasets or infinite sequences because they generate values only as needed.

Defining Generators

      1.Using the yield keyword inside a function.
      2.Generator Expressions

```python
#generator using yield
def countdown(n):
    while n > 0:
        yield n
        n -= 1

gen = countdown(5)
for num in gen:
    print(num)
```

```python
#generator using expression
gen = (x * x for x in range(5))

print(next(gen))
print(next(gen))
print(next(gen))
```

# **7. What are the advantages of using generators over regular functions?**

**Ans 7.**

Here are the key advantages of generators over regular functions:

  1. Memory Efficiency:

    Generators produce values one at a time using the yield keyword, meaning they do not store the entire result in memory

  2. Lazy Evaluation:

    Generators perform lazy evaluation, meaning values are computed only when requested. This allows you to start processing data immediately without waiting for the entire result to be available.

  3. Improved Performance:

    Since generators yield values one at a time, they avoid the overhead of creating large lists or sequences all at once.

  4. Suitability for Infinite Sequences:

    Generators are ideal for generating infinite sequences because they yield values one by one and can continue indefinitely without exhausting memory.

  5. Avoiding Intermediate Data Structures:

    Generators avoid the creation of intermediate data structures like lists or sets, reducing the overhead of storing temporary results.



```python
#example of generator

def lazy_range(n):
    for i in range(n):
        yield i


gen = lazy_range(10)
print(next(gen))
print(next(gen))
```

# **9. Explain the purpose and usage of the `map()` function in Python.**

**Ans 9.**

The map() function in Python is used to apply a given function to all the items of an iterable (like a list, tuple, etc.) and return a map object (an iterator) with the results. The map() function is especially useful when you want to transform or process each element in an iterable without using an explicit loop.

    Syntax:
    map(function, iterable, ...)

```python
#example of map()
def square(x):
    return x * x

numbers = [1, 2, 3, 4, 5]
squared_numbers = map(square, numbers)
print(list(squared_numbers))
```

# **10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?**

Ans 10.

The map(), reduce(), and filter() functions in Python are all functional programming tools that operate on iterables. While they are similar in that they apply a function to the elements of an iterable, they differ significantly in how they operate and the results they return

    map(): Transforms each element of an iterable by applying a function.

    filter(): Filters elements of an iterable based on a condition (function that returns True or False).

    reduce(): Reduces an iterable to a single cumulative value by applying a function repeatedly.



```python
#example of map
numbers = [1, 2, 3, 4]
squared_numbers = map(lambda x: x * x, numbers)
print(list(squared_numbers))
```

```python
#example of filter()
numbers = [1, 2, 3, 4, 5, 6]
even_numbers = filter(lambda x: x % 2 == 0, numbers)
print(list(even_numbers))
```

```python
#example of reduce()
from functools import reduce

numbers = [1, 2, 3, 4, 5]
sum_of_numbers = reduce(lambda x, y: x + y, numbers)
print(sum_of_numbers)
```
